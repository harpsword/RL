```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
def draw(algo, seeds, env_name):
    plt.xlabel("episode")
    plt.ylabel("performance")
    for seed in seeds:
        path = os.path.join(algo, algo+'-'+env_name+"-seed-"+str(seed)+"--reward.csv")
        results = pd.read_csv(path, index_col = 0)
        y = results['0']
        x = np.arange(1, len(y)+1)
        plt.plot(x[::], y[::],linewidth=0.8)
    plt.title(env_name)
    plt.show()
```

```python

plt.rcParams['savefig.dpi'] = 250 #图片像素
plt.rcParams['figure.dpi'] = 250 #分辨率
```

```python
algo = "ddpg"
envs_list = ['HalfCheetah-v2', 'Hopper-v2', 'Humanoid-v2',
             'HumanoidStandup-v2','InvertedDoublePendulum-v2',
             'Swimmer-v2', 'Walker2d-v2', 'InvertedPendulum-v2', 'Reacher-v2']
seeds = [199862, 217042, 415717, 455745, 683513]
```

```python

for env_name in envs_list:
    draw(algo, seeds, env_name)
```
