```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
def draw(folder, algo, seeds, env_name):
    plt.xlabel("episode")
    plt.ylabel("performance")
    for seed in seeds:
        path = os.path.join(folder, algo+'-'+env_name+"-seed-"+str(seed)+"--reward.csv.npy")
        imagepath = os.path.join(folder, algo+'-'+env_name+"--reward.png")
        y = np.load(path)
        x = np.arange(1, len(y)+1)
        plt.plot(x[::], y[::],linewidth=0.8)
    plt.title(env_name+"-"+algo)
    
    plt.savefig(imagepath)
    plt.show()
```

```python

plt.rcParams['savefig.dpi'] = 250 #图片像素
plt.rcParams['figure.dpi'] = 250 #分辨率
```

```python
folder = "TD3R"
algo = 'DDPG'
envs_list = ['HalfCheetah-v2', 'Hopper-v2', 'Humanoid-v2',
             'HumanoidStandup-v2','InvertedDoublePendulum-v2',
             'Swimmer-v2', 'Walker2d-v2', 'InvertedPendulum-v2', 'Reacher-v2']
seeds = [12806, 54834, 158683, 555798, 679754]
```

```python
for env_name in envs_list:
    draw(folder, 'TD3', seeds, env_name)
```

```python

for env_name in envs_list:
    draw(folder, algo, seeds, env_name)
```
