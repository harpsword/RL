```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#results = pd.read_csv("ppo-record-Amidar-v0.csv", index_col=0)
results = pd.read_csv("ppo-record-Assault-v0.csv", index_col=0)
```

```python
count = {
    
}

gamelength = {}
```

```python
for i in range(results.shape[0]):
    episode = results.iloc[i, 0]
    reward = results.iloc[i, 2]
    if episode not in count.keys() and episode != 0:
        count[episode] = [reward]
    elif episode in count.keys():
        count[episode].append(reward)
        
```

```python
x = []
y = []
for k, v in count.items():
    x.append(k)
    y.append(np.mean(v))
```

```python
x = np.array(x)
y = np.array(y)
sort_index = np.argsort(x)
x = x[sort_index]
y = y[sort_index]
```

```python

plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率
# plt.rcParams['figure.figsize'] = (4.0, 2.0) # 设置figure_size尺寸
```

```python
plt.plot(x[::30], y[::30],linewidth=0.8)
plt.xlabel("episode")
plt.ylabel("performance")
# plt.show()

# plt.savefig('Qbert.png', dpi=300)
```
