```python
import gym
import torch
import numpy as np
from matplotlib import pyplot as plt
```

```python
model = MLPPolicy(4, 2)
```

```python
# print('test:', test_model(model, env))
```

```python
env = gym.make('CartPole-v0')
```

```python
def test_model(model, env):
    obs = env.reset()
    done = False
    R = 0
    while not done:
        action = model.return_action(obs)
        obs, Reward, done, _ = env.step(action) 
        R += Reward

    return R
```

```python
generation_list = []
test_Reward_list = []

Generation = 2000
one_time = 10
i = 0
while i < Generation:
    model = train(model, one_time, lr=0.00005)
    i += one_time
    generation_list.append(i)
    test_Reward_list.append(test_model(model, env))
```

```python
plt.plot(generation_list, test_Reward_list)
```
